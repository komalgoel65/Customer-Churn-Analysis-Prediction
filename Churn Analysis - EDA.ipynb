```python

telco_data = telco_base_data.copy()
```

```python
#Removing missing values 
telco_data.dropna(how = 'any', inplace = True)

#telco_data.fillna(0)
```

```python

# Get the max tenure
print(telco_data['tenure'].max()) #72
```

```python

# Group the tenure in bins of 12 months
labels = ["{0} - {1}".format(i, i + 11) for i in range(1, 72, 12)]

telco_data['tenure_group'] = pd.cut(telco_data.tenure, range(1, 80, 12), right=False, labels=labels)
```

```python
telco_data['tenure_group'].value_counts()
```

```python
#drop column customerID and tenure
telco_data.drop(columns= ['customerID','tenure'], axis=1, inplace=True)
telco_data.head()
```

UNIVARIATE ANALYSIS

```python
import matplotlib.pyplot as plt
import seaborn as sns

for i, predictor in enumerate(telco_data.drop(columns=['Churn', 'TotalCharges', 'MonthlyCharges'])):
    plt.figure(i, figsize=(6,4))
    sns.countplot(data=telco_data, x=predictor, hue='Churn')
    plt.title(f"Countplot of {predictor} vs Churn")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()   # ðŸ‘ˆ ensures each plot is displayed
```

```python

telco_data['Churn'] = np.where(telco_data.Churn == 'Yes',1,0)
```

```python
telco_data.head()
```

```python
telco_data_dummies = pd.get_dummies(telco_data)
telco_data_dummies.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Make sure both are numeric
telco_data_dummies['MonthlyCharges'] = pd.to_numeric(telco_data_dummies['MonthlyCharges'], errors='coerce')
telco_data_dummies['TotalCharges'] = pd.to_numeric(telco_data_dummies['TotalCharges'], errors='coerce')

# Plot
sns.lmplot(data=telco_data_dummies, x='MonthlyCharges', y='TotalCharges', fit_reg=False)
plt.title("Monthly Charges vs Total Charges")  
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))

Mth = sns.kdeplot(
    telco_data_dummies.loc[telco_data_dummies["Churn"] == 0, "MonthlyCharges"],
    color="red", fill=True, label="No Churn"
)

Mth = sns.kdeplot(
    telco_data_dummies.loc[telco_data_dummies["Churn"] == 1, "MonthlyCharges"],
    color="blue", fill=True, label="Churn", ax=Mth
)

Mth.set_ylabel('Density')
Mth.set_xlabel('Monthly Charges')
Mth.set_title('Monthly Charges by Churn')
Mth.legend(loc='upper right')

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))

Tot = sns.kdeplot(
    telco_data_dummies.loc[telco_data_dummies["Churn"] == 0, "TotalCharges"],
    color="red", fill=True, label="No Churn"
)

Tot = sns.kdeplot(
    telco_data_dummies.loc[telco_data_dummies["Churn"] == 1, "TotalCharges"],
    color="blue", fill=True, label="Churn", ax=Tot
)

Tot.set_ylabel('Density')
Tot.set_xlabel('Total Charges')
Tot.set_title('Total Charges by Churn')
Tot.legend(loc='upper right')

plt.show()
```

```python

new_df1_target0=telco_data.loc[telco_data["Churn"]==0]
new_df1_target1=telco_data.loc[telco_data["Churn"]==1]
```

```python
def uniplot(df, col, title, hue=None):
    import matplotlib.pyplot as plt
    import seaborn as sns

    sns.set_style('whitegrid')
    sns.set_context('talk')
    plt.rcParams["axes.labelsize"] = 14
    plt.rcParams['axes.titlesize'] = 18
    plt.rcParams['axes.titlepad'] = 20
    
    if hue is not None:
        temp = df[hue]
        hue_count = len(temp.unique())
    else:
        hue_count = 0
    
    fig, ax = plt.subplots()
    width = max(8, len(df[col].unique()) + 4 + 2*hue_count)  # adjusted width
    fig.set_size_inches(width , 6)
    
    plt.xticks(rotation=45)
    plt.title(title)
    
    ax = sns.countplot(data=df, x=col, hue=hue, order=df[col].value_counts().index, palette='bright')
    
    plt.show()
```

```python
# For Churned customers
uniplot(new_df1_target1, col='Partner', title='Distribution of Gender for Churned Customers', hue='gender')

# For Non-Churned customers
uniplot(new_df1_target0, col='Partner', title='Distribution of Gender for Non Churned Customers', hue='gender')
```

```python
uniplot(new_df1_target1,col='PaymentMethod',title='Distribution of PaymentMethod for Churned Customers',hue='gender')
```

```python

uniplot(new_df1_target1,col='Contract',title='Distribution of Contract for Churned Customers',hue='gender')
```
